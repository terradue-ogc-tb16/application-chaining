```python
import subprocess
import graphviz
```

```python
def run_command(command, **kwargs):
    
    process = subprocess.Popen(args=command, stdout=subprocess.PIPE, **kwargs)
        
    return process.stdout.read()
```

```python
command = ['cwltool',
           '--print-dot',
           'burned-area.cwl#burned-area']
```

```python
dot_content = run_command(command).decode()
```

```python
graphviz.Source(dot_content)
```

```python
graphviz.Source(dot_content).render('burned-area', view=True, format='png')
```
